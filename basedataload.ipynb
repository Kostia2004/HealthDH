```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import os
```

```python
#      
image = None
mask = None
minx = 9999
miny = 9999
maxx = 0
maxy = 0
count = 0
for path in os.listdir('./tissue-segment/'):
    if '_mask' not in path:
        image = cv.imread('./tissue-segment/'+path, cv.IMREAD_GRAYSCALE)
        mask = cv.imread('./tissue-segment/'+path.split('.')[0]+'_mask.jpg', cv.IMREAD_GRAYSCALE)
        if image.shape[0]>maxx:
            maxx = image.shape[0]
        if image.shape[1]>maxy:
            maxy = image.shape[1]
        if image.shape[0]<minx:
            minx = image.shape[0]
        if image.shape[1]<miny:
            miny = image.shape[1]
        count+=1
        
print(minx, miny)
print(maxx, maxy)
print(count)
fig = plt.figure(figsize=(40, 20))
plt.subplot(1,2,1)
plt.imshow(image, cmap='bone')
plt.subplot(1,2,2)
plt.imshow(mask)
```
